```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
data=pd.read_csv('1.01. Simple linear regression.csv')
```

```python
data
```

```python
data.describe()
```

```python
# y=c+bx
y=data['GPA']
x=data['SAT']
plt.scatter(x,y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

```python
x1=sm.add_constant(x)
results=sm.OLS(y,x1).fit()
results.summary()
```

```python
plt.scatter(x,y)
yhat= x*0.0017+ 0.275 # constants are taken from summary 
fig=plt.plot(x,yhat, lw=4,c='orange',label='regression line')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()   
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
data=pd.read_csv('real_estate_price_size.csv')
```

```python
data
```

```python
y=data['price']
x=data['size']
plt.scatter(x,y)
plt.xlabel('price', fontsize=20)
plt.ylabel('size', fontsize=20)
plt.show()
```

```python
x1=sm.add_constant(x)
results=sm.OLS(y,x1).fit()
results.summary()
```

```python
plt.scatter(x,y)
yhat=x*223.1787+1.019e+05 # constants are taken from summary 
fig=plt.plot(x,yhat, lw=4,c='orange',label='regression line')
plt.xlabel('price',fontsize=20)
plt.ylabel('size', fontsize=20)
plt.show()
```
